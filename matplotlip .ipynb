```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig,ax=plt.subplots()            # create a fig contining a single axes
ax.plot([1,2,3,4,],[4,3,1,2]);   # plot same data on axes
```

```python
x=[1,2,3,4,]
y=[4,3,1,2]
```

```python
plt.figure(figsize=(6,2)) # it is used for size (should be in 1 step)
plt.plot(x,y)             # value
plt.ylabel('y_axis')      # y_ axis lable
plt.xlabel('x_axis')      # x_ axis lable 
plt.title('sample plot')  # title
plt.show()                # to show the image 
```
